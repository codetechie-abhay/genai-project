# **LANGGRAPH**

‚úÖ Step 4: Define State (memory of graph)

LangGraph works with a state dictionary.



````python
from typing import TypedDict

class State(TypedDict):
    topic: str
    explanation: str
````

‚úÖ Step 5: Create Nodes (functions = tasks)

Each node is a task.

````python


def explain_node(state: State):
    response = llm.invoke(f"Explain {state['topic']} in short.")
    return {"explanation": response.content}


def interview_node(state: State):
    response = llm.invoke(
        f"Create 3 interview questions from this explanation:\n{state['explanation']}"
    )
    return {"explanation": response.content}
````

‚úÖ Step 6: Build the Graph
from langgraph.graph import StateGraph, END


````python

from langgraph.graph import StateGraph, END

graph = StateGraph(State)

graph.add_node("explain", explain_node)
graph.add_node("interview", interview_node)

graph.set_entry_point("explain")
graph.add_edge("explain", "interview")
graph.add_edge("interview", END)

app = graph.compile()
````

‚úÖ Step 7: Run LangGraph

````python

result = app.invoke({"topic": "pauri garhwal"})
print(result)
````

# Customer Support Chatbot with LangGraph Conditional Flow Architecture

## Overview
This notebook demonstrates a complete customer support chatbot built using LangGraph's conditional flow state architecture. The system follows a structured workflow:

```
__start__ ‚Üí categorize ‚Üí analyze_sentiment ‚Üí conditional_routing ‚Üí handler_functions ‚Üí __end__
```

## Architecture Components:
1. **State Management**: Maintains query, category, sentiment, and response
2. **Categorization**: Classifies queries into billing, technical, general, or escalate
3. **Sentiment Analysis**: Analyzes emotional tone (positive, neutral, negative)
4. **Conditional Routing**: Intelligently routes based on category and sentiment
5. **Handler Functions**: Specialized responses for each category
6. **Escalation**: Automatically escalates sensitive/angry queries

## Dependencies:
- LangGraph for workflow orchestration
- LangChain for LLM integration
- Groq for fast inference
- Python-dotenv for environment management

````python
from typing_extensions import TypedDict

from langchain_core.runnables import RunnableLambda
from langchain_core.prompts import ChatPromptTemplate

from langchain_groq import ChatGroq

from langgraph.graph import StateGraph, END

from IPython.display import Image, display
````

````python

from langgraph.graph import StateGraph
from typing_extensions import TypedDict
class State(TypedDict):
    """
    State structure for the customer support chatbot.
    This dictionary maintains all information as it flows through the workflow.

    Attributes:
        query (str): The original customer query
        category (str): Categorized as billing, technical, general, or escalate
        sentiment (str): Analyzed as positive, neutral, or negative
        response (str): The final response generated for the customer
    """
    query: str
    category: str
    sentiment: str
    response: str

# Initialize the LLM with Groq
llm = ChatGroq(temperature=0, model_name="llama-3.1-8b-instant")

print("üèóÔ∏è State structure defined and LLM initialized!")


````

# **‚úÖ Step 2: Create categorize node (customer care)**

````python
def categorize(state: State) -> State:
    """
    Categorize the user's query into billing, technical, general, or escalate.
    This is the first processing step in our workflow.

    Args:
        state (State): Current workflow state containing the customer query

    Returns:
        State: Updated state with category information
    """
    print("üîç ---CATEGORIZING---")

    # Create a prompt for categorization
    prompt = ChatPromptTemplate.from_messages([
        ("system", "You are a customer support agent. Categorize the user's query into one of the following categories: billing, technical, general, escalate. If the user is angry or the query is sensitive, categorize it as 'escalate'. Respond with only the category name."),
        ("human", "{query}")
    ])

    # Create the processing chain
    category_chain = prompt | llm | StrOutputParser()

    # Process the query
    category = category_chain.invoke({"query": state["query"]})
    print(f"üìã Category: {category}")

    return {"category": category.strip().lower()}

def analyze_sentiment(state: State) -> State:
    """
    Analyze the sentiment of the user's query.
    This happens after categorization and helps determine routing.

    Args:
        state (State): Current workflow state

    Returns:
        State: Updated state with sentiment information
    """
    print("üòä ---ANALYZING SENTIMENT---")

    # Create a prompt for sentiment analysis
    prompt = ChatPromptTemplate.from_messages([
        ("system", "You are a sentiment analysis expert. Analyze the sentiment of the user's query and categorize it as: positive, neutral, negative. Respond with only the sentiment."),
        ("human", "{query}")
    ])

    # Create the processing chain
    sentiment_chain = prompt | llm | StrOutputParser()

    # Process the query
    sentiment = sentiment_chain.invoke({"query": state["query"]})
    print(f"üí≠ Sentiment: {sentiment}")

    return {"sentiment": sentiment.strip().lower()}

print("‚úÖ Core processing functions defined!")
````

````python
def handle_billing(state: State) -> State:
    """
    Handle billing-related queries with specialized responses.

    Args:
        state (State): Current workflow state

    Returns:
        State: Updated state with billing response
    """
    print("üí∞ ---HANDLING BILLING QUERY---")

    prompt = ChatPromptTemplate.from_messages([
        ("system", "You are a billing support specialist. Provide helpful and professional responses to billing-related queries. Be clear, concise, and empathetic."),
        ("human", "{query}")
    ])

    response_chain = prompt | llm | StrOutputParser()
    response = response_chain.invoke({"query": state["query"]})
    print(f"üí≥ Billing Response: {response}")

    return {"response": response}

def handle_technical(state: State) -> State:
    """
    Handle technical support queries with detailed solutions.

    Args:
        state (State): Current workflow state

    Returns:
        State: Updated state with technical response
    """
    print("üîß ---HANDLING TECHNICAL QUERY---")

    prompt = ChatPromptTemplate.from_messages([
        ("system", "You are a technical support specialist. Provide helpful and professional responses to technical queries. Be clear, provide step-by-step solutions when possible, and be empathetic."),
        ("human", "{query}")
    ])

    response_chain = prompt | llm | StrOutputParser()
    response = response_chain.invoke({"query": state["query"]})
    print(f"‚öôÔ∏è Technical Response: {response}")

    return {"response": response}

def handle_general(state: State) -> State:
    """
    Handle general customer service queries.

    Args:
        state (State): Current workflow state

    Returns:
        State: Updated state with general response
    """
    print("üìû ---HANDLING GENERAL QUERY---")

    prompt = ChatPromptTemplate.from_messages([
        ("system", "You are a customer service representative. Provide helpful and professional responses to general queries. Be friendly, informative, and empathetic."),
        ("human", "{query}")
    ])

    response_chain = prompt | llm | StrOutputParser()
    response = response_chain.invoke({"query": state["query"]})
    print(f"ü§ù General Response: {response}")

    return {"response": response}

def escalate(state: State) -> State:
    """
    Escalate sensitive or angry queries to human agents.
    This is used for negative sentiment or sensitive issues.

    Args:
        state (State): Current workflow state

    Returns:
        State: Updated state with escalation response
    """
    print("üö® ---ESCALATING TO HUMAN AGENT---")

    response = "I understand your concern and I'm sorry for any inconvenience. This query requires special attention, and I'm escalating it to our human support team who will be able to assist you better. A representative will contact you shortly."
    print(f"üìû Escalation Response: {response}")

    return {"response": response}

print("‚úÖ Handler functions defined!")
````

````python
def route_query(state: State) -> str:
    """
    Route the query based on category and sentiment analysis.
    This is the decision point in our conditional flow architecture.

    Routing Logic:
    - If sentiment is negative OR category is escalate ‚Üí escalate
    - If category is billing ‚Üí handle_billing
    - If category is technical ‚Üí handle_technical
    - Otherwise ‚Üí handle_general

    Args:
        state (State): Current workflow state with category and sentiment

    Returns:
        str: Name of the next node to route to
    """
    print("üéØ ---ROUTING QUERY---")

    category = state.get("category", "").lower()
    sentiment = state.get("sentiment", "").lower()

    # Priority routing: negative sentiment or escalate category goes to escalation
    if sentiment == "negative" or category == "escalate":
        print("üö® Routing to: ESCALATE")
        return "escalate"
    elif category == "billing":
        print("üí∞ Routing to: BILLING")
        return "handle_billing"
    elif category == "technical":
        print("üîß Routing to: TECHNICAL")
        return "handle_technical"
    else:
        print("üìû Routing to: GENERAL")
        return "handle_general"

print("‚úÖ Conditional routing function defined!")
````

````python
def create_customer_support_graph():
    """
    Create the complete customer support chatbot workflow graph.
    This function builds the entire LangGraph architecture.

    Returns:
        StateGraph: Compiled workflow ready for execution
    """
    print("üèóÔ∏è Building Customer Support Graph...")

    # Create the graph with our State structure
    workflow = StateGraph(State)

    # Add all processing nodes
    workflow.add_node("categorize", categorize)
    workflow.add_node("analyze_sentiment", analyze_sentiment)
    workflow.add_node("handle_billing", handle_billing)
    workflow.add_node("handle_technical", handle_technical)
    workflow.add_node("handle_general", handle_general)
    workflow.add_node("escalate", escalate)

    # Define the workflow edges

    # 1. Start ‚Üí categorize ‚Üí analyze_sentiment (fixed sequence)
    workflow.add_edge("categorize", "analyze_sentiment")

    # 2. Conditional routing from analyze_sentiment to handlers
    workflow.add_conditional_edges(
        "analyze_sentiment",  # Source node
        route_query,           # Routing function
        {                      # Mapping of routing decisions to target nodes
            "handle_billing": "handle_billing",
            "handle_technical": "handle_technical",
            "handle_general": "handle_general",
            "escalate": "escalate"
        }
    )

    # 3. All handler nodes lead to END
    workflow.add_edge("handle_billing", END)
    workflow.add_edge("handle_technical", END)
    workflow.add_edge("handle_general", END)
    workflow.add_edge("escalate", END)

    # 4. Set the entry point of the workflow
    workflow.set_entry_point("categorize")

    # 5. Compile the graph into an executable application
    app = workflow.compile()

    print("‚úÖ Customer Support Graph created successfully!")
    return app

# Create the compiled graph
customer_support_app = create_customer_support_graph()
print("üöÄ Graph is ready for use!")
````

````python
def run_customer_support(query: str) -> Dict[str, str]:
    """
    Process a customer query through the complete LangGraph workflow.
    This is the main interface function for our chatbot.

    Args:
        query (str): The customer's query/question

    Returns:
        Dict[str, str]: Complete results including category, sentiment, and response
    """
    print("="*80)
    print(f"üîç PROCESSING QUERY: {query}")
    print("="*80)

    # Process the query through the workflow
    results = customer_support_app.invoke({"query": query})

    # Display final results
    print("="*80)
    print("üìã FINAL RESULT:")
    print(f"üìù Query: {results['query']}")
    print(f"üìÇ Category: {results['category']}")
    print(f"üòä Sentiment: {results['sentiment']}")
    print(f"üí¨ Response: {results['response']}")
    print("="*80)

    return results

print("‚úÖ Main interface function ready!")
````
